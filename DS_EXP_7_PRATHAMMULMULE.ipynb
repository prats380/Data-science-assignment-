```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_openml
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
```

```python
iris = fetch_openml(name='iris', version=1)
X, y = pd.DataFrame(iris.data), pd.Series(iris.target)
```

```python
print(X.head())
print(y.head())
```

```python
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Selected features using ANOVA F-value:", selected_features.tolist())
```

```python
mi_selector = SelectKBest(score_func=mutual_info_classif, k=2)
X_mi_new = mi_selector.fit_transform(X, y)
mi_selected_features = X.columns[mi_selector.get_support()]
print("Selected features using Mutual Information:", mi_selected_features.tolist())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.bar(X.columns, selector.scores_)
plt.title('Feature Scores from ANOVA F-value')
plt.xticks(rotation=45)
plt.show()
```
